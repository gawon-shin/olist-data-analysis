```python
import pandas as pd

# ===============================
# Q3. 배송일 구간 생성 (Pandas)
# ===============================
# SQL에서 계산된 delivery_days를 기준으로
# 배송 기간을 3개 구간으로 분류하여
# Power BI에서 분포 분석에 활용

# 1. 데이터 로드
file_path = "../data/q3_delivery_time_base.csv"
df = pd.read_csv(file_path)

# 2. 배송 구간 생성
bins = [-1, 5, 10, 20, float("inf")]
labels = ["5일 이하", "6~10일", "11~20일", "20일 초과"]

df["delivery_bucket"] = pd.cut(
    df["delivery_days"],
    bins=bins,
    labels=labels
)

df.head()

# 3. 결과 저장
output_path = "../data/q3_delivery_time_pandas.csv"
df.to_csv(output_path, index=False, encoding="utf-8-sig")
```
